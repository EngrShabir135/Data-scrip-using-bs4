```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.scrapethissite.com/pages/simple/"
```

```python
req=requests.get(url)
```

```python
print(req)
```

```python
soup=BeautifulSoup(req.text,"html.parser")
```

```python
print(soup.prettify)
```

```python
name=soup.find("h3",{"class":"country-name"}).text.strip()
```

```python
name
```

```python
Contry_name=[]
for contry_na in soup.find_all("h3",{"class":"country-name"}):
    con_name=contry_na.get_text(strip=True)
    Contry_name.append(con_name)
```

```python
len(Contry_name)
```

```python
Capital=[]
for cap in soup.find_all("span",{"class":"country-capital"}):
    capi=cap.get_text(strip=True)
    Capital.append(capi)
```

```python
print(len(Capital))
```

```python
Population=[]
for popu in soup.find_all("span",{"class":"country-population"}):
    popula=popu.get_text(strip=True)
    Population.append(popula)
```

```python
len(Population)
```

```python
Area=[]
for area in soup.find_all("span",{"class":"country-area"}):
    Ar=area.get_text(strip=True)
    Area.append(Ar)
```

```python
len(Area)
```

```python
data={"Country_name":Contry_name,
     "Capital":Capital,
     "Countr_population":Population,
     "Area":Area}
```

```python
df=pd.DataFrame(data=data)
```
